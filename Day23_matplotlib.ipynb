```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
xs =np.linspace(0,50,100)
ys = np.sin(xs)

# print(xs)
# print(ys)
figure, axe = plt.subplots()
figure.suptitle('Figure1')
figure.supylabel('Figure_y')
figure.supxlabel('Figure_x')
figure.text(0.5,0.7, 'mytext')
figure.set_size_inches(15,4.8)
figure.set_facecolor('red') #       

axe.plot(xs, ys)
axe.set_title('axe_title1')
axe.set_xlabel('axe_x')
axe.set_ylabel('axe_y')
axe.set_xlim(0,30)
axe.set_ylim(0,1)
axe.set_xticks(np.linspace(0, 30, 5))
axe.set_xticklabels(list('abcde'))
print(type(figure))
print(type(axe))
```
