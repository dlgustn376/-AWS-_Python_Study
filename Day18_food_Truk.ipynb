## 전국푸드트럭허가구역 표준데이타

- 전국 / 지자체 푸드트럭 허가구역 은 갯수
- 전국 / 지자체별 푸드트럭 운영대수

```python
import numpy as np
import pandas as pd

pop_df = pd.read_csv('전국푸드트럭허가구역표준데이터.csv',encoding='cp949')
pop_df.drop(0,axis=0, inplace=True)  # inplace 작업은 하면 할 수록 데이터를 날림.

# print(pop_df.iloc[0])

print(pop_df.iloc[:,2].count())
print(pop_df.iloc[:,9].count())
pop_df.columns
```

```python
print(pop_df['푸드트럭운영대수'].value_counts())
```

```python
sub_data = pop_df[['시도명','푸드트럭운영대수']]

sub_data['푸드트럭운영대수'] = sub_data['푸드트럭운영대수'].replace(np.NaN, 0)
print(sub_data)
```
