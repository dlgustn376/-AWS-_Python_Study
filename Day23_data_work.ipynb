### 데이터산업의 기술등급별 데이터직무 인력 현황
* 데이터 직무별 axe 생성
    - x축 : 시점
    - y축 : 업종 2개
* 하나의 figure에, 총 8개 axe가 바둑판 형식으로 배치되는 형태 (4 col x 2 rows), figure 사이즈는 각자 알아서.
* title 표시

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()

matplotlib.rcParams['font.family'] = [kfont]

# fig, axe = plt.subplots()
# fig.suptitle('한글abcde')
```

```python
pd_data = pd.read_csv('데이터산업의_기술등급별_데이터직무_인력_현황_2018-2021.csv', encoding='cp949',header=0)
# col_selection = ['데이터직무별','시점', '데이터 처리 및 관리 솔루션 개발·공급업', '데이터 구축 및 컨설팅 서비스업' ]
# pd_data = pd_data.drop(['기술등급별(1)'], axis=1)
pd_data
```

```python
i=7
print(pd_data.iloc[4*i:4*(i+1)-3,0])
```

```python
for i in range(8):
    job = pd_data.iloc[4*i:4*(i+1)-3,0]
    filter1 = pd_data['데이터직무별(1)'] == str(job)
    sub_data = pd_data[filter1]
    
    fig,axe = plt.subplots(2,4)
    fig.set_figwidth(6.4*3)
    axe[i//4,i%4].plot(sub_data['시점'], sub_data.iloc[4*i:4*(i+1),3])
    axe[i//4,i%4].plot(sub_data['시점'], sub_data.iloc[4*i:4*(i+1),4])
```

```python
filter1 = pd_data['데이터직무별(1)'] =='데이터 아키텍트'
sub_data = pd_data[filter1]

a,axe = plt.subplots()
a.suptitle('데이터 아키텍트')
axe.set_xlabel('시점')
axe.set_ylabel('인구현황')
axe.plot(sub_data['시점'], sub_data.iloc[:,3])
axe.plot(sub_data['시점'], sub_data.iloc[:,4])
```
