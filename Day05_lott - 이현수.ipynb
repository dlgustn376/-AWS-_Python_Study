```python
csv_data=[]
with open('lott.CSV') as f:
    for line in f:
        csv_data.append(line[0:-1].split(','))
        #csv_data.append(line.replace('\n','').split(','))
print(csv_data[:3])
int(csv_data[0][-1])
```

```python
import math
# 1. 전체 회차를 통틀어 빈도수 (mode): 숫자별 카운트 (몇번이나 나왔나..) 1-45
#mode_list = [int(e) for e in csv_data for e in e]
#print(mode_list)
data_list = []
for e in csv_data:
    for i, e_i in enumerate(e):
        e[i] = int(e_i)
        data_list.append(e[i])
#print(data_list)

mode_value = set(mode_list)
print(mode_value)
def count_mode(mode_list):
    count_mode = {}
    for i in mode_list:
        if i in count_mode:
            count_mode[i] += 1
        else:
            count_mode[i] = 1
    return count_mode
mode = count_mode(data_list)

kmode = mode.keys()
vmode = mode.values()
print(mode)
#print(kmode)
#print(vmode)
```

```python
# 2. 회차별 - 평균값, 표준편차


#sum_value = 0
#for i in range(0,len(num_list)+1):
#    for j, e in enumerate(num_list[i]):
#         = int(e_i)
#        sum_value += e
#print(sum_value)
num_list = []
for i in csv_data:
    num_list.append(i)
print(num_list[1])

sum_value = 0
for i in num_list:
    for index, number in enumerate(i):
        i[index]= int(number)
        sum_value += i[index]
    print(sum_value/7)
    sum_value = 0


def mean(num_list):
    sum_value=0
    for i in num_list:
        sum_value += i
    return sum_value/len(num_list)

def dev(num_list):
    m = mean(num_list)
    return [x-m for x in num_list]

def var(num_list):
    n = len(num_list)
    d = dev(num_list)
    return sum([x*x for x in d])/(n-1)

def stdev(num_list):
    return math.sqrt(var(num_list))

#print()
#print(mean(kmode))
#print(mean(vmode))

#print()
#print(dev(kmode))
#print(dev(vmode))

#print()
#print(var(kmode))
#print(var(vmode))

#print()
print('표준편차 key: ', stdev(kmode))
print('표준편차 values: ', stdev(vmode))
```
