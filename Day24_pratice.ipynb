```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family'] = [kfont]

pd_data = pd.read_csv('데이터산업_시장규모_2015-2021.csv', encoding='cp949')
print(pd_data)
```

```python
fig = plt.figure(layout='tight', figsize=(6.4*3,4.8*3))
line_axe = fig.add_subplot(2,1,1)
col_names = pd_data.columns
line_params = ['ro-','g^--','bv:']
label = ['데이터 처리 및 관리 솔루션 개발·공급업','데이터 구축 및 컨설팅 서비스업','데이터 판매 및 제공 서비스업']

# line_axe.set_title('2015-2021 산업별 규모')
# line_axe.set_ylabel('억원')
line_axe.set(title='2015-2021 산업별 규모', ylabel='억원')

for i in range(1,4):
#     line_axe.plot(pd_data['시점'], pd_data[col_names[i]], line_params[i-1], linewidth=1)
    line_axe.plot('시점', col_names[i],  line_params[i-1], linewidth=3, data=pd_data)
    
    
pie2015_axe = fig.add_subplot(2,3,4)
pie2015_axe.set_title('2015년 구성비', y=-0.2)
pie2015_axe.pie(pd_data.iloc[0,1:],radius=1, autopct='%.2f', labels=label)

pie2021_axe = fig.add_subplot(2,3,5)
pie2021_axe.set_title('2021년 구성비', y=-0.2)
pie2021_axe.pie(pd_data.iloc[-1,1:], radius=1 , autopct='%.2f', labels=label)

fig.legend(loc="lower right")
```
