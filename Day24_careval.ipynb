```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data'
pd_data = pd.read_csv(url, header=None)
pd_data.columns = ['구매가', '유지비용', '문갯수','탑승인원', '수납공간', '안전도', '만족도']
print(pd_data.shape)
print(pd_data.dtypes)
pd_data.head()
```

* 구매가가 만족도에 미치는 영향.
    - figure를 좌우로 나우어, 왼쪽은 구매가 vs. 만족도 산점도, 오른쪽은 아래위로 분할, 위쪽은 구매가 vhigh  중 만족도별 count, 아래쪽은 구매가 low중 만족도별 count
* 수납공간이 크면 탑승인원도 많은가.

```python
fig = plt.figure(layout='tight', figsize=(6.4*3,4.8*3))
line_axe = fig.add_subplot(1,2,1)
vhigh_axe = fig.add_subplot(2,2,2)
low_axe = fig.add_subplot(2,2,4)
col_names = pd_data.columns

for e in pd_data['구매가'].unique():
    filter1 = pd_data['구매가']==e
    xs = pd_data.loc[filter1,'구매가']
    ys = pd_data.loc[filter1, '만족도']
    line_axe.scatter(xs,ys)


# vhigh_axe.bar()
```
