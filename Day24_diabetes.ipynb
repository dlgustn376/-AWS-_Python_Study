```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn import datasets

url = 'https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt'
pd_data = pd.read_csv(url, sep='\t')

print(pd_data)
```

```python
_,axe = plt.subplots()
axe.plot(pd_data['BMI'])
```

```python
_,axe = plt.subplots()
# axe.hist(pd_data['BMI'], bins=10, linewidth=1.0, edgecolor="white")
# axe.hist(pd_data['BP'], bins=10, linewidth=1.0, edgecolor="white")
axe.hist(pd_data['AGE'], bins=10, linewidth=1.0, edgecolor="white")
```

```python
_,axe = plt.subplots()
axe.boxplot(pd_data['AGE'])
```

```python
_, axe = plt.subplots()
axe.pie(pd_data['SEX'].value_counts(), radius=1, center=(4, 4),
        wedgeprops={"linewidth": 1, "edgecolor": "white"})
```
