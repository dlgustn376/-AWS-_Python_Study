### 데이터산업의 기술등급별 데이터직무 인력 현황
* 데이터 직무별 axe 생성
    - x축 : 시점
    - y축 : 업종 2개
* 하나의 figure에, 총 8개 axe가 바둑판 형식으로 배치되는 형태 (4 col x 2 rows), figure 사이즈는 각자 알아서.
* title 표시

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# current_font_list = matplotlib.rcParams
font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family'] = [kfont]

pd_data = pd.read_csv('데이터산업의_기술등급별_데이터직무_인력_현황_2018-2021.csv', encoding='cp949')
print(pd_data.head())
```

```python
fig, axes = plt.subplots(2,4, figsize=(6.4*4,4.8*2), sharey=True, sharex=True)
col_names = pd_data.columns[-2:]

for i, n in enumerate(pd_data['데이터직무별(1)'].unique()):
    filter1 = pd_data['데이터직무별(1)'] ==n
    ax = axes[i//4, i%4]
    xs = pd_data.loc[filter1, '시점']
    ys = pd_data.loc[filter1, col_names]
    ax.set_title(n)
#     ax.plot(xs.astype('str'),ys.iloc[:,0] , 'go--')
    ax.plot(xs.astype('str'),ys.iloc[:,0] , 'gv--')
    ax.plot(xs.astype('str'),ys.iloc[:,1] , 'ro--')
```

```python
filter1 = pd_data['데이터직무별(1)'] =='데이터 아키텍트'
col_names = pd_data.columns[-2:]
xs = pd_data.loc[filter1, '시점']
ys = pd_data.loc[filter1, col_names]
_,axe = plt.subplots()
axe.plot(xs,ys )
```
